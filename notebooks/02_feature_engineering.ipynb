# Model Selection and Tuning Process

## Feature Engineering
- **Numerical Features**: age, trestbps, chol, thalach, oldpeak
  - Applied median imputation for missing values
  - Standardized using StandardScaler (mean=0, std=1)
  
- **Categorical Features**: sex, cp, fbs, restecg, exang, slope, ca, thal
  - Applied most frequent imputation for missing values
  - One-hot encoded for model compatibility

## Model Selection
Two classification models were selected:
1. **Logistic Regression**: Linear model for binary classification, interpretable, good baseline
2. **Random Forest**: Ensemble method, handles non-linear relationships, robust to outliers

## Training Process
- Train/test split: 80/20 with stratification
- Cross-validation: 5-fold CV on training set
- Evaluation metrics: Accuracy, Precision, Recall, ROC-AUC

## Hyperparameter Tuning
Used GridSearchCV with 5-fold cross-validation:

**Logistic Regression Parameters:**
- C (regularization strength): [0.001, 0.01, 0.1, 1, 10, 100]
- penalty: ['l2']
- solver: ['liblinear', 'lbfgs']

**Random Forest Parameters:**
- n_estimators: [50, 100, 200]
- max_depth: [None, 10, 20, 30]
- min_samples_split: [2, 5, 10]
- min_samples_leaf: [1, 2, 4]

## Model Evaluation
Models evaluated on test set using:
- Accuracy: Overall correctness
- Precision: True positive rate among predicted positives
- Recall: True positive rate among actual positives
- ROC-AUC: Area under ROC curve (ranking performance)
